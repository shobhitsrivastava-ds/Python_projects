```python
import numpy as np
import math
#Function to calculate the hcf of the two numbers :
def hcf(x, y):  
   if(x > y):  
       smaller = y  
   else:  
       smaller = x  
   for i in range(1,smaller+1):  
       if((x % i == 0) and (y % i == 0)):  
           hcf1 = i  
   return hcf1

#Function to find the lcm of the two number :
def lcm(x, y):  
   if x > y:  
       greater = x  
   else:  
       greater = y  
   while(True):  
       if((greater % x == 0) and (greater % y == 0)):  
           lcm = greater  
           break  
       greater += 1  
   return lcm  

#main function for performing the operation
def smart_calculator():
    text=[]
    while(1):
        text =input("Enter some operation ")
        length =len(text)
        num= [i for i in text.split(" ") if i.isdigit()]
        let=[l for l in text.split(" ")]
        #print(let)
        #performing the sum operation
        for k in text.split(" "):
            # Sum
            if(k=="sum" or k== "addition" or k== "add"):
                print(int(num[0]) + int(num[1]))
            # Substraction
            elif(k=="substract" or k=="minus"):
                print(int(num[0]) - int(num[1]))
            # Multiplication
            elif(k=="multiply" or k=="mul"):
                print(int(num[0]) * int(num[1]))
            # Division
            elif(k=="divide" or k=="div"):
                print(int(num[0]) / int(num[1]))
            # HCF
            elif(k=="hcf"):
                print(hcf( int(num[0]),int(num[1])))
            # LCM
            elif(k=="lcm"):
                print(lcm(int(num[0]),int(num[1])))
            # MODULO
            elif(k=="modulus" or k=="modulo"):
                print(int(num[0]) % int(num[1]))
            # SIN
            elif(k=="sin"):
                print(math.sin(int(num[0])))
            # COS
            elif(k=="cos"):
                print(math.cos(int(num[0])))
            # POWER
            elif(k=="pow" or k== "power"):
                print(math.pow(int(num[0]),int(num[1])))
            # SQUARE ROOT
            elif(k == "square" or k=="root"):
                print(math.sqrt(int(num[0])))
            # CUBE
            elif(k == "cube"):
                print(math.pow(int(num[0]),3))
            #EXIT
            elif(k=="exit"):
                print("Thank you ! visit again")
        
smart_calculator()
```
